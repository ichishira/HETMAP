```python
sample_id = "TCGA-AO-A03L-01Z-00-DX1"
print(sample_id)
slide_path = f"/wsi/analysis/TCGA_TMA_analysis/data/WSI/{cancer_type}/{sample_id}.svs"
print(slide_path)
slide = openslide.OpenSlide(slide_path)
try:
    level = slide.get_best_level_for_downsample(128)
    # 画像全体のサイズを取得
    width, height = slide.level_dimensions[level]

    # OpenSlideで画像を読み込む
    img = slide.read_region((0, 0), level, (width, height))
    
    # 解像度情報を表示
    #print("スライド名:", slide_path)
    #print("レベル数:", slide.level_count)
    #print("各レベルのサイズ:")
    #for i in range(slide.level_count):
    #    print(f"  レベル {i}: {slide.level_dimensions[i]}")
    
    # マイクロン/ピクセル単位の解像度情報
    if "openslide.mpp-x" in slide.properties:
        mpp_x = float(slide.properties["openslide.mpp-x"])
        mpp_y = float(slide.properties["openslide.mpp-y"])
        print(f"\n解像度: {mpp_x} µm/px (X), {mpp_y} µm/px (Y)")
        if mpp_x<0.3 and mpp_y<0.3:
            print("mag40")
        else:
            print("mag20")
    else:
        print("\n解像度情報が取得できませんでした。")
        non_mag_info_sample.append(sample_id)
except:
    broken_sample.append(sample_id)
    print(f"{sample_id} is broken!")
```

```python
mag_20_sample =[]
mag_40_sample = []
non_mag_info_sample = []
broken_sample=[]

for sample_id in sample_with_clone_segpath: 
    print(sample_id)
    slide_path = f"/wsi/analysis/TCGA_TMA_analysis/data/WSI/{cancer_type}/{sample_id}.svs"
    print(slide_path)
    slide = openslide.OpenSlide(slide_path)
    try:
        level = slide.get_best_level_for_downsample(128)
        # 画像全体のサイズを取得
        width, height = slide.level_dimensions[level]
    
        # OpenSlideで画像を読み込む
        img = slide.read_region((0, 0), level, (width, height))
        
        # 解像度情報を表示
        #print("スライド名:", slide_path)
        #print("レベル数:", slide.level_count)
        #print("各レベルのサイズ:")
        #for i in range(slide.level_count):
        #    print(f"  レベル {i}: {slide.level_dimensions[i]}")
        
        # マイクロン/ピクセル単位の解像度情報
        if "openslide.mpp-x" in slide.properties:
            mpp_x = float(slide.properties["openslide.mpp-x"])
            mpp_y = float(slide.properties["openslide.mpp-y"])
            print(f"\n解像度: {mpp_x} µm/px (X), {mpp_y} µm/px (Y)")
            if mpp_x<0.3 and mpp_y<0.3:
                mag_40_sample.append(sample_id)
            else:
                mag_20_sample.append(sample_id)
        else:
            print("\n解像度情報が取得できませんでした。")
            non_mag_info_sample.append(sample_id)
    except:
        broken_sample.append(sample_id)
        print(f"{sample_id} is broken!")
```
